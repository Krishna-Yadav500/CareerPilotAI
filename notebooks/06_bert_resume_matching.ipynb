```python
from sentence_transformers import SentenceTransformer

from sklearn.metrics.pairwise import cosine_similarity

import pandas as pd
```

```python
model = SentenceTransformer(
    'all-MiniLM-L6-v2'
)
```

```python
resume_text = """

Python
Machine Learning
Deep Learning
TensorFlow
Pandas
NumPy
Scikit-Learn
NLP
Computer Vision
Git

"""
```

```python
jobs = [

"""
Machine Learning Intern

Python
Machine Learning
TensorFlow
Deep Learning
Pandas
NumPy
""",

"""
Frontend Developer

React
JavaScript
HTML
CSS
""",

"""
Data Scientist

Python
Statistics
Machine Learning
SQL
Pandas
""",

"""
Cyber Security Analyst

Network Security
Penetration Testing
Linux
""",

"""
AI Engineer

Deep Learning
Computer Vision
PyTorch
TensorFlow
"""
]
```

```python
resume_embedding = model.encode(
    resume_text
)
```

```python
job_embeddings = model.encode(
    jobs
)
```

```python
scores = cosine_similarity(

    [resume_embedding],

    job_embeddings

)
```

```python
scores
```

```python
job_roles = [

    "ML Intern",

    "Frontend Developer",

    "Data Scientist",

    "Cyber Security Analyst",

    "AI Engineer"
]
```

```python
results = pd.DataFrame({

    "Role":job_roles,

    "Similarity Score":
    scores[0] * 100
})
```

```python
results = results.sort_values(

    by="Similarity Score",

    ascending=False
)
```

```python
results
```

```python
best_role = results.iloc[0]

print(
    "Best Career Match:"
)

print(
    best_role["Role"]
)

print(
    round(
        best_role[
            "Similarity Score"
        ],
        2
    )
)
```

```python
results.to_csv(

    "../outputs/bert_job_matches.csv",

    index=False
)

print(
    "Saved Successfully"
)
```
