```python
import pdfplumber
import re
import pandas as pd
```

```python
def extract_text_from_pdf(pdf_path):
    
    text = ""

    with pdfplumber.open(pdf_path) as pdf:

        for page in pdf.pages:

            text += page.extract_text() + "\n"

    return text
```

```python
resume_text = extract_text_from_pdf(
     r"C:\Users\raurk\OneDrive\Documents\Desktop\Prime\CareerPilotAI\notebooks\KrishnaYadavMLResume.pdf"
)

print(resume_text[:3000])
```

```python
def extract_email(text):

    pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}'

    match = re.search(pattern, text)

    if match:
        return match.group()

    return None
```

```python
email = extract_email(resume_text)

print(email)
```

```python
def extract_phone(text):

    pattern = r'(\+91[- ]?)?[6-9]\d{9}'

    match = re.search(pattern, text)

    if match:
        return match.group()

    return None
```

```python
phone = extract_phone(resume_text)

print(phone)
```

```python
def extract_name(text):

    lines = text.split('\n')

    return lines[0]
```

```python
name = extract_name(resume_text)

print(name)
```

```python
skills_db = [

    "python",
    "java",
    "c++",
    "sql",
    "html",
    "css",
    "javascript",
    "machine learning",
    "deep learning",
    "tensorflow",
    "pytorch",
    "pandas",
    "numpy",
    "scikit-learn",
    "streamlit",
    "git",
    "opencv"
]
```

```python
def extract_skills(text):

    found_skills = []

    text = text.lower()

    for skill in skills_db:

        if skill.lower() in text:

            found_skills.append(skill)

    return found_skills
```

```python
skills = extract_skills(resume_text)

print(skills)
```

```python
candidate = {

    "Name": name,
    "Email": email,
    "Phone": phone,
    "Skills": skills
}

candidate
```

```python
df = pd.DataFrame([candidate])

df.to_csv(
    r"C:\Users\raurk\OneDrive\Documents\Desktop\Prime\CareerPilotAI\outputs\candidate_profile.csv",
    index=False
)

print("Saved Successfully")
```
