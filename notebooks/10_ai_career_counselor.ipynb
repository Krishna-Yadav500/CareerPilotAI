```python
import google.generativeai as genai
```

```python

# import streamlit as st

# genai.configure(
#     api_key=st.secrets["GEMINI_API_KEY"]
# )
```

```python
model = genai.GenerativeModel(
    "models/gemini-2.5-flash"
)
```

```python
student_profile = """

Name: Krishna Yadav

Education:
B.Tech 3rd Year

Skills:
Python
NumPy
Pandas
Machine Learning

Goal:
Get ML Internship

Timeline:
6 Months

"""
```

```python
target_role = "Machine Learning Engineer"
```

```python
prompt = f"""
You are an experienced AI Career Counselor.

Student Profile:

{student_profile}

Target Role:

{target_role}

Generate:

1. Career Readiness Score

2. Missing Skills

3. 6-Month Learning Roadmap

4. Recommended Projects

5. Certifications

6. Interview Preparation Plan

7. Weekly Study Schedule

8. Internship Preparation Tips

Provide detailed guidance.
"""
```

```python
response = model.generate_content(
    prompt
)

print(response.text)
```

```python
with open(
    "../outputs/career_counseling_report.txt",
    "w",
    encoding="utf-8"
) as f:

    f.write(response.text)

print("Report Saved Successfully")
```
