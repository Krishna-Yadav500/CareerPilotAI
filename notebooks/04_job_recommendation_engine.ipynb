```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
jobs = [

    {
        "Role":"Machine Learning Intern",

        "Description":
        """
        Python
        Machine Learning
        Deep Learning
        TensorFlow
        Pandas
        NumPy
        Scikit-Learn
        """
    },

    {
        "Role":"Data Analyst",

        "Description":
        """
        SQL
        Excel
        Power BI
        Python
        Pandas
        Data Analysis
        """
    },

    {
        "Role":"Software Engineer",

        "Description":
        """
        Java
        C++
        Data Structures
        Algorithms
        Git
        """
    },

    {
        "Role":"AI Engineer",

        "Description":
        """
        Python
        Deep Learning
        TensorFlow
        PyTorch
        NLP
        Computer Vision
        """
    },

    {
        "Role":"Data Scientist",

        "Description":
        """
        Python
        SQL
        Machine Learning
        Statistics
        Pandas
        NumPy
        """
    }
]
```

```python
jobs_df = pd.DataFrame(jobs)

jobs_df
```

```python
resume_skills = [

    "python",
    "machine learning",
    "numpy",
    "pandas",
    "git"
]
```

```python
resume_profile = " ".join(resume_skills)

print(resume_profile)
```

```python
corpus = [resume_profile]

corpus.extend(
    jobs_df["Description"].tolist()
)

len(corpus)
```

```python
vectorizer = TfidfVectorizer()

vectors = vectorizer.fit_transform(
    corpus
)
```

```python
resume_vector = vectors[0]

job_vectors = vectors[1:]

similarities = cosine_similarity(
    resume_vector,
    job_vectors
)
```

```python
jobs_df["Match Score"] = similarities[0] * 100

jobs_df
```

```python
recommended_jobs = jobs_df.sort_values(
    by="Match Score",
    ascending=False
)

recommended_jobs
```

```python
top_jobs = recommended_jobs.head(5)

top_jobs
```

```python
for i,row in top_jobs.iterrows():

    print("="*50)

    print("Role:", row["Role"])

    print(
        "Match Score:",
        round(
            row["Match Score"],
            2
        ),
        "%"
    )

    print("="*50)
```

```python
def recommendation_label(score):

    if score >= 80:
        return "Highly Recommended"

    elif score >= 60:
        return "Recommended"

    elif score >= 40:
        return "Consider Learning"

    else:
        return "Low Match"
```

```python
jobs_df["Recommendation"] = jobs_df[
    "Match Score"
].apply(
    recommendation_label
)

jobs_df
```

```python
jobs_df.to_csv(
    "../outputs/job_recommendations.csv",
    index=False
)

print(
    "Recommendations Saved"
)
```

```python
best_job = jobs_df.sort_values(
    by="Match Score",
    ascending=False
).iloc[0]

print(
    "Best Career Match:"
)

print(
    best_job["Role"]
)

print(
    round(
        best_job["Match Score"],
        2
    ),
    "%"
)
```

```python
def missing_skills_for_job(
    job_description,
    resume_skills
):

    job_skills = job_description.lower().split()

    missing = []

    for skill in job_skills:

        if skill not in resume_skills:

            missing.append(skill)

    return list(set(missing))
```

```python
jobs_df["Missing Skills"] = jobs_df[
    "Description"
].apply(

    lambda x:
    missing_skills_for_job(
        x,
        resume_skills
    )
)

jobs_df[
    [
        "Role",
        "Missing Skills"
    ]
]
```
