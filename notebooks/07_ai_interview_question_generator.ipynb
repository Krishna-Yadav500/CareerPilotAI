```python
import google.generativeai as genai
```

```python


# import streamlit as st

# genai.configure(
#     api_key=st.secrets["GEMINI_API_KEY"]
# )
```

```python
model = genai.GenerativeModel(
    "models/gemini-2.5-flash"
)
```

```python
resume_text = """
Python
Machine Learning
Deep Learning
TensorFlow
NLP
Computer Vision
Projects:
CareerPilot AI
"""
```

```python
target_role = "Machine Learning Intern"
```

```python
prompt = f"""
You are an experienced technical interviewer.

Candidate Resume:

{resume_text}

Target Role:

{target_role}

Generate:

1. 5 Technical Questions
2. 3 Project Questions
3. 2 HR Questions

Return in a clean format.
"""
```

```python
response = model.generate_content(
    prompt
)

print(response.text)
```

```python
with open(
    "../outputs/interview_questions.txt",
    "w",
    encoding="utf-8"
) as f:

    f.write(response.text)

print("Saved Successfully")
```
