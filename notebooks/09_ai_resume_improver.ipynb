```python
model = genai.GenerativeModel(
    "models/gemini-2.5-flash"
)
```

```python
resume_text = """
Krishna Yadav

Skills:
Python, Machine Learning, Pandas, NumPy

Projects:
CareerPilot AI

Education:
B.Tech

Experience:
Fresher
"""
```

```python
prompt = f"""
You are an expert ATS resume reviewer.

Analyze this resume:

{resume_text}

Provide:

1. ATS Score (0-100)

2. Resume Strengths

3. Resume Weaknesses

4. Missing Skills

5. Improved Professional Summary

6. Improved Project Description

7. Resume Improvement Suggestions

Return in professional format.
"""
```

```python
response = model.generate_content(
    prompt
)

print(response.text)
```

```python
with open(
    "../outputs/resume_improvement_report.txt",
    "w",
    encoding="utf-8"
) as f:

    f.write(response.text)

print("Resume Report Saved")
```
