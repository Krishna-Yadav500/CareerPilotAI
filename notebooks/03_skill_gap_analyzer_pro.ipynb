```python
import pandas as pd
```

```python
resume_skills = [
    "python",
    "numpy",
    "pandas",
    "machine learning",
    "git"
]

jd_skills = [
    "python",
    "numpy",
    "pandas",
    "machine learning",
    "deep learning",
    "tensorflow",
    "sql",
    "streamlit",
    "git"
]
```

```python
skill_categories = {

    "Programming":[
        "python",
        "java",
        "c++",
        "javascript"
    ],

    "ML":[
        "machine learning",
        "deep learning",
        "tensorflow",
        "pytorch",
        "scikit-learn"
    ],

    "Data":[
        "numpy",
        "pandas",
        "sql",
        "data analysis"
    ],

    "Tools":[
        "git",
        "streamlit",
        "docker",
        "github"
    ]
}
```

```python
missing_skills = list(
    set(jd_skills) -
    set(resume_skills)
)

print(missing_skills)
```

```python
skill_gap = {}

for category, skills in skill_categories.items():

    gap = []

    for skill in missing_skills:

        if skill in skills:

            gap.append(skill)

    if gap:

        skill_gap[category] = gap

skill_gap
```

```python
priority = {

    "python":10,
    "machine learning":10,
    "deep learning":9,
    "tensorflow":9,
    "sql":8,
    "streamlit":7,
    "git":6
}
```

```python
ranked_skills = sorted(
    missing_skills,
    key=lambda x: priority.get(x,1),
    reverse=True
)

ranked_skills
```

```python
learning_resources = {

    "deep learning":
    "Deep Learning Specialization",

    "tensorflow":
    "TensorFlow Developer Course",

    "sql":
    "SQL for Data Science",

    "streamlit":
    "Streamlit Crash Course",

    "python":
    "Python for Everybody"
}
```

```python
recommendations = {}

for skill in ranked_skills:

    recommendations[skill] = \
        learning_resources.get(
            skill,
            "Self Study"
        )

recommendations
```

```python
readiness = (
    len(resume_skills)
    /
    len(jd_skills)
) * 100

print(
    f"Career Readiness Score: "
    f"{readiness:.2f}%"
)
```

```python
if readiness >= 85:
    label = "Job Ready"

elif readiness >= 70:
    label = "Almost Ready"

elif readiness >= 50:
    label = "Needs Upskilling"

else:
    label = "Beginner"

print(label)
```

```python
roadmap = []

for skill in ranked_skills:

    roadmap.append({

        "Skill":skill,

        "Resource":
        recommendations[skill]
    })

roadmap
```

```python
roadmap_df = pd.DataFrame(
    roadmap
)

roadmap_df
```

```python
roadmap_df.to_csv(
    "../outputs/learning_roadmap.csv",
    index=False
)

print(
    "Roadmap Saved Successfully"
)
```

```python
report = {

    "Career Readiness":
    round(readiness,2),

    "Status":
    label,

    "Missing Skills":
    missing_skills,

    "Priority Skills":
    ranked_skills
}

report
```
